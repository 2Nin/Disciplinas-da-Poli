```python
"""
     ***EP 1 - PMR3201***
Nome: Antônio Augusto Carnevalli
N°USP: 13682909

"""

# O problema serve para ilustrar conceitos de programação dinâmica.
# A ideia consiste em ir criando problemas sucessivos a cada iteração,
# verificando todas as possibilidades de subsequências. 
# A tabela solução diz qual é a maior subsequência (pode haver mais de uma).


def maior_subsequencia(a, b):

  # Verificando se há listas vazias
  if (len(a) ==0 or len(b) == 0) or (len(a) == 0 and len(b) == 0):
    print('Foi dado uma lista vazia')
    return 0

  # Garantindo que a é a maior lista
  if len(a) < len(b):
    a, b = b, a
  a = list(a)  # Separando a lista em elementos individuais
  b = list(b)

  # Criando a matriz de zeros
  L = [[0 for coluna in range(len(b)+1)]for linha in range(len(a)+1)]

  # Substituindo a matriz de zeros pela tabela de solução
  for i in range(len(a)+1):
    for j in range(len(b)+1):
        if i == 0 or j == 0:
            L[i][j] = 0
        elif a[i-1] == b[j-1]:
            L[i][j] = L[i-1][j-1] + 1
        else:
            L[i][j] = max(L[i-1][j], L[i][j-1])


  # Iteraremos a tabela solução para achar a maior subsequência
  # Não poderemos usar um for nesse caso, pois precisaremos que
  # o algorítimo 'pule' iterações quando ocorrer a ordem da maior subsequência
  maiorsubseq = []
  i = len(a)
  j = len(b)
  while i>0 and j>0:
    if a[i-1] == b[j-1]:
      maiorsubseq.append(a[i-1])
      i-=1
      j-=1

    elif L[i-1][j] > L[i][j-1]:
      i -= 1
             
    else:
      j -= 1

  # Como percorremos a tabela de trás para frente, precisaremos inverter a subsequência
  maiorsubseq = maiorsubseq[::-1]
  return (''.join(maiorsubseq)), len(maiorsubseq)

# Função que faz o tratamento de arquivos txt, dada no enunciado do EP
def LeArquivoDNA(filename):

    files=open(filename, 'r')
    lists=files.readlines()  # matriz de n linhas
    nlines=len(lists)  # numero de linhas
    a = lists[0].rstrip('\n').split(' ')  # separa a linhas em colunas
    # observa o espaco ’ ’ como
    # caracter de separacao
    # descarta \n
    cadeiacompleta = a[1] + a[2] + a[3] + a[4] + a[5] + a[6]
    # concatena as colunas 1..6 ignora a[0]
    # agora que cadeiacompleta nao e’vazio faca ate o final
    for i in range(1, nlines):
        a=lists[i].rstrip('\n').split(' ')
        cadeiacompleta = cadeiacompleta + a[1]+a[2]+a[3]+a[4]+a[5]+a[6]
    # retorna a string completa, e o nome do arquivo
    return cadeiacompleta, filename


# Criando as variáveis de teste

x1 = 'AGGTAB'
y1 = 'GXTXAYB'

x2 = 'ATGGGTGATGTTGAGAAAGGCAAGAAGATTTTTATTATGAAGTGTTCCCAGTGCCACACC'
y2 = 'ATGGGTGATGTTGAGAAAGGCAAGAAGATTTTTATTATGAAGTGTTCCCAGTGCCATACC'

x3 = LeArquivoDNA('DengueVirus2StrainBA05i_Jakarta.txt')
y3 = LeArquivoDNA('DengueVirus3StrainTB55i_KualaLumpur.txt')

x4 = LeArquivoDNA('InfluenzaTypeA_H1N1_California.txt')
y4 = LeArquivoDNA('InfluenzaTypeA_H3N2_NewYork.txt')

x5 = LeArquivoDNA('SARS-COV2-Spike_Toronto2.txt')
y5 = LeArquivoDNA('SARS-COV2-Spike_Wuhan1.txt')

# Gerando o arquivo dos resultados
nomedoarquivo = input('Digite o nome do arquivo a ser criado com os resultados:\n')
with open(f'{nomedoarquivo}.txt', 'w') as f:
    # Para as funções geradas sem arquivos
    f.write(f'\nTeste 1\n'
            f'Primeira Sequência = {x1}\n'
            f'Segunda Sequência = {y1}\n'
            f'A maior subsequência em comum é:\n{maior_subsequencia(x1, y1)[0]}, de tamanho {maior_subsequencia(x1, y1)[1]}.\n')
    f.write(f'\nTeste 2\n'
            f'Primeira Sequência = {x2}\n'
            f'Segunda Sequência = {y2}\n'
            f'A maior subsequência em comum é:\n{maior_subsequencia(x2, y2)[0]}, de tamanho {maior_subsequencia(x2, y2)[1]}.\n')

    # Para as funções geradas com arquivos
    f.write(f'\nTeste 3\n'
            f'A primeira sequência vem do arquivo = {x3[1]}\n'
            f'A segunda sequência vem do arquivo = {y3[1]}\n'
            f'A maior subsequência em comum é:\n{maior_subsequencia(x3[0], y3[0])[0]}\nDe tamanho {maior_subsequencia(x3[0], y3[0])[1]}.\n')
    f.write(f'\nTeste 4\n'
            f'A primeira sequência vem do arquivo = {x4[1]}\n'
            f'A segunda sequência vem do arquivo = {y4[1]}\n'
            f'A maior subsequência em comum é:\n{maior_subsequencia(x4[0], y4[0])[0]}\nDe tamanho {maior_subsequencia(x4[0], y4[0])[1]}.\n')
    f.write(f'\nTeste 5\n'
            f'A primeira sequência vem do arquivo = {x5[1]}\n'
            f'A segunda sequência vem do arquivo = {y5[1]}\n'
            f'A maior subsequência em comum é:\n{maior_subsequencia(x5[0], y5[0])[0]}\nDe tamanho {maior_subsequencia(x5[0], y5[0])[1]}.\n')

# Imprimindo o conteúdo do arquivo no console
with open(f'{nomedoarquivo}.txt', 'r') as f:
    print(f.read())
```
